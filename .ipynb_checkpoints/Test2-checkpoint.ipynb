```python
data=[40,48,50,45,35,65,70,95]
import numpy as np
exam_score=np.array(data)
mean_score=np.mean(exam_score)
median_score=np.median(exam_score)
std_deviation=np.std(exam_score)
variance=np.var(exam_score)
print('Mean:',mean_score)
print('Median:',median_score)
print('Standard Deviation:', std_deviation)
print('Variance:', variance)
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randint(0,101,size=100)
bins=[0,25,50,75,100]
plt.hist(data,bins=bins,color='blue',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
def calculate_bmi(height, weight):
    height_meters = height / 100 
    bmi = weight / (height_meters ** 2)
    
   
    if bmi < 18.5:
        status = 'Underweight'
    elif 18.5 <= bmi < 24.9:
        status = 'Normal Weight'
    elif 25 <= bmi < 29.9:
        status = 'Overweight'
    else:
        status = 'Obese'
    
    return bmi, status


height = 175 
weight = 95   
bmi_result, health_status = calculate_bmi(height, weight)
print(f"BMI: {bmi_result:.2f}, Health Status: {health_status}")
```
